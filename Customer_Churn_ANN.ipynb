```python
#converting to numerical to object
df['TotalCharges'].isnull().sum()
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.dropna()
df.drop('customerID',inplace=True,axis=1)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Churn'] = le.fit_transform(df['Churn'])
df['gender'] = le
```

```python
# Ensure x and y are correctly defined from df_encoded before splitting
# This step prevents inconsistencies if previous cells were not run in order
from sklearn.model_selection import train_test_split
x = df_encoded.drop("Churn",axis=1)
y = df_encoded['Churn']

# Identify and drop rows with NaN values from x and y consistently
nan_rows = x.isnull().any(axis=1)
x = x[~nan_rows]
y = y[~nan_rows]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)

#scalling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) # Scale the training features
x_test = sc.transform(x_test)       # Scale the test features

from imblearn.over_sampling import SMOTE
smt = SMOTE()
# Apply SMOTE to the training data to balance classes, reassigning to x_train and y_train
x_train, y_train = smt.fit_resample(x_train, y_train)
print(x_train.shape,y_train.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import models,layers

model = models.Sequential([
    layers.Dense(128,activation='relu',input_shape=(x_train.shape[1],)),
    layers.Dropout(0.2),
    layers.Dense(64,activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1,activation='sigmoid')


])
model.compile(optimizer='adam',

              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=32,validation_data=(x_test,y_test))
```

```python
loss , accuracy = model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
orignal_y = (y_test > 0.5)
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(orignal_y,y_pred))
print(classification_report(orignal_y,y_pred))

```
